## 1) ¿Qué es una variable?
Una variable es un nombre que almacena un valor en memoria.

```python
edad = 20
altura = 1.75
nombre = "Ana"
es_estudiante = True

print("Edad:", edad)
print("Altura:", altura)
print("Nombre:", nombre)
print("¿Es estudiante?:", es_estudiante)
```

Defina sus propias variables de un problema ambiental e imprimalas con un mensaje

```python
saludo = "hola"
print(saludo, "mundo")
```

## 2) Entrada de datos
`input()` siempre devuelve texto (string).

```python
nombre_usuario = input("Escribe tu nombre: ")
print("Hola,", nombre_usuario)
```

```python
edad_usuario = int(input("Escribe tu edad: "))
print("En 5 años tendrás:", edad_usuario + 5)
```

## 3) Nombres de variables
- No pueden empezar con número
- No pueden tener espacios
- Python distingue mayúsculas
- Estilo recomendado: snake_case

```python
nombre_completo = "Diego Fernando Ramírez Rodríguez"
NombreCompleto = "Ejemplo CamelCase"

print(nombre_completo)
print(NombreCompleto)
```

## 4) Comentarios

## 6) Tipos de dato

```python
a = 10
b = 3.14
c = "hola"
d = False

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

## 7) Operadores aritméticos

```python
x = 10
y = 3

print("Suma:", x + y)
print("Resta:", x - y)
print("Multiplicación:", x * y)
print("División:", x / y)
print("División entera:", x // y)
print("Residuo:", x % y)
print("Potencia:", x ** y)
```

## 8) Ejemplo integrador
### distancia = velocidad * tiempo

```python
velocidad = float(input("Velocidad (m/s): "))
tiempo = float(input("Tiempo (s): "))

distancia = velocidad * tiempo
print(f"La distancia recorrida es {distancia} metros")
```
