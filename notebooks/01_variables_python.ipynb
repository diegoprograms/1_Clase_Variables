## 1) ¿Qué es una variable?
Una **variable** es un nombre que referencia un valor.

En Python, el tipo de dato lo determina el valor asignado.

```python
edad = 20
altura = 1.72
nombre = "Ana"
es_estudiante = True

print("Edad:", edad)
print("Altura:", altura)
print("Nombre:", nombre)
print("¿Es estudiante?:", es_estudiante)
```

## 2) Salida de variables: `print()`
La función `print()` muestra información en pantalla.

```python
x = 10
y = 3
print("x =", x)
print("y =", y)
```

## 3) Entrada de variables: `input()`
**Importante:** `input()` siempre devuelve un **string**. Por eso a veces hay que convertir a `int` o `float`.

```python
nombre_usuario = input("Escribe tu nombre: ")
print("Hola,", nombre_usuario)
```

```python
edad_usuario = int(input("Escribe tu edad (entero): "))
print("En 5 años tendrás:", edad_usuario + 5)
```

## 4) Nombres de variables
Reglas rápidas:
- No pueden empezar con número: ❌ `1dato`
- No pueden tener espacios: ❌ `nombre completo`
- Evitar caracteres especiales: ❌ `nombre-completo`
- Python distingue mayúsculas: `edad` ≠ `Edad`

Estilos:
- `nombre_completo` (snake_case) ✅ recomendado en Python
- `NombreCompleto` (CamelCase) se usa más para clases

```python
nombre_completo = "Diego Fernando Ramírez Rodríguez"
NombreCompleto = "Ejemplo CamelCase"

print(nombre_completo)
print(NombreCompleto)
```

## 5) Comentarios
- Comentario de una línea: `# ...`
- Bloque (string multilínea): `""" ... """` (más usado como documentación)

```python
# Esto es un comentario de una línea

"""
Esto es un bloque de texto multilínea.
Suele usarse como documentación.
"""

a = 5  # comentario al final
print(a)
```

## 6) Palabras reservadas (keywords)
Son palabras que Python ya usa (no se pueden usar como nombres de variable):
`if`, `else`, `for`, `while`, `def`, `return`, `class`, `True`, `False`, `None`, `import`, etc.

```python
import keyword
keyword.kwlist
```

## 7) Tipos de dato básicos
- `int`: enteros
- `float`: decimales
- `str`: texto
- `bool`: verdadero/falso

```python
i = 10
f = 3.14
s = "hola"
b = False

print(i, type(i))
print(f, type(f))
print(s, type(s))
print(b, type(b))
```

## 8) Operadores aritméticos
- `+` suma, `-` resta, `*` multiplicación, `/` división
- `//` división entera, `%` módulo (residuo), `**` potencia

```python
x = 10
y = 3

print("x + y =", x + y)
print("x - y =", x - y)
print("x * y =", x * y)
print("x / y =", x / y)
print("x // y =", x // y)
print("x % y =", x % y)
print("x ** y =", x ** y)
```

## 9) Ejemplo integrador: `distancia = velocidad * tiempo`

```python
velocidad = float(input("Velocidad (m/s): "))
tiempo = float(input("Tiempo (s): "))

distancia = velocidad * tiempo
print(f"Si v={velocidad} m/s y t={tiempo} s, entonces d={distancia} m")
```
